```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn import datasets
import random
import warnings
warnings.filterwarnings("ignore")
```

```python
boston = datasets.load_boston()
boston
```

```python
features = pd.DataFrame(boston.data, columns=boston.feature_names)
features.head()
```

```python
target = boston.target
df = features
df["Target"] = target
df.head()
```

```python
# check for null values
def check_null(df):
    for column in df.columns:
        print(f"{column} has {df[column].isnull().sum()} null values")

check_null(df)
```

```python
# visulaize the distributions of each of the variables
colors = ["red", "green", "blue", "yellow", "cyan"]
color = random.choice(colors)
plt.figure(figsize=(10, 20))

sns.displot(data=df,x="INDUS",y=None,hue=None,
row=None,col=None,weights=None,kind='hist',
rug=False,rug_kws=None,log_scale=None,
legend=True,palette=None,hue_order=None,
hue_norm=None,color=None,col_wrap=None,
row_order=None,col_order=None,height=5,
aspect=1,facet_kws=None)
    
plt.show()
```

```python
# plot correlation
def plot_coorelation(df, feature, color):
    plt.scatter(df[feature], df["Target"], color=color)
    plt.title(f"{feature} vs Target", c="green")
    plt.xlabel(feature, c="yellow")
    plt.ylabel("Target", c="yellow")
    
features = df.columns
length = len(features)
colors = ["red", "cyan", "pink", "grey"]
plt.figure(figsize=(20, 15))
for i in range(length - 1):
    color = random.choice(colors)
    feature = features[i]
    plt.subplot(7, 2, i + 1)
    plot_coorelation(df, feature, color)
```

```python
x = df.drop("Target", axis=1)
y = df.Target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
print(f"{x_train.shape}, {x_test.shape}")
```

```python
ran_for = RandomForestRegressor()
ran_for.fit(x_train, y_train)
```

```python
ran_for.score(x_train, y_train)
```

```python
ran_for.score(x_test, y_test)
```

```python
# evaluating the base model perfomance
def plot_prediction_curves(model, x_train, x_test, y_train, y_test):
    train_predict = model.predict(x_train)
    test_predict = model.predict(x_test)
    plt.plot([y_train.min(), y_train.max()], [y_train.min(),y_train.max()],color="green", label="train")
    plt.plot([y_test.min(), y_test.max()], [y_test.min(),y_test.max()],color="red", label="test")
    plt.scatter(y_train, train_predict, color="green", marker="+")
    plt.scatter(y_test, test_predict, color="red", marker="+")
    plt.title("Prediction Curves")
    plt.legend()

plt.figure(figsize=(35, 15))
plt.style.use('fivethirtyeight')
plot_prediction_curves(ran_for, x_train, x_test, y_train, y_test)
```

```python
# perform hyperparameter tuning
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [5, 10, 15, 20, 25],
    'min_samples_split': [2, 4, 6, 8, 10],
    'min_samples_leaf': [2, 4, 6, 8, 10],
}

grid = GridSearchCV(ran_for, param_grid, cv=5, n_jobs=-1, verbose=2)
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
# plot the learning curves
from sklearn.metrics import mean_squared_error
def plot_learning_curves(model, x, y):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    train_error, test_error = [], []
    for m in range(1, len(x_train)):
        model.fit(x_train[:m], y_train[:m])
        y_train_predict = model.predict(x_train[:m])
        y_test_predict = model.predict(x_test)
        train_error.append(mean_squared_error(y_train[:m], y_train_predict))
        test_error.append(mean_squared_error(y_test, y_test_predict))
        
    plt.plot(np.sqrt(train_error), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(test_error), "b-", linewidth=3, label="test")
    plt.legend()
        
        

plot_learning_curves(final_model, x, y)
```
